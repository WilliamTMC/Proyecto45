Para comenzar el análisis, se me ocurrió la brillante idea de visitar 30 mipymes en mi provincia y recoger en cada una de ellas información sobre 10 productos. Fue una tarea bastante tediosa, ya que muchas personas pensaban que yo era inspector, y se armaban tremendo lío en sus cabezas. En muchos casos, para evitar molestias y malentendidos, siempre pregunté respetuosamente si podía o no recopilar información en cada mipyme que encontraba. Finalmente,cuando logré reunir todo lo necesario,se me ocurrió analizar los productos que más repetidos entre mipymes. Seleccioné los 10 más frecuentes y a partir de ellos, realicé un análisis sobre su disponibilidad y el valor promedio que tienen en la provincia.

```python
from Funciones.biblioteca import open_json,obtener_nombres_y_precios
import glob
import matplotlib.pyplot as plt
import operator
#Creo una variable para almacenar todo el contenido de la carpeta Mipymes
direcciones = glob.glob("./Mipymes/*.json")#utilizo glob.glob para de la carpeta Mipymes me lea todos los .json
#Creo una lista para guardar todo el contenido almacenado en direcciones
contenido = []
for i in direcciones :
    contenido.append(dict(open_json(i)))#llamamos a la funcion open_json para que nos lea cada uno de los json
    
#Creo una variable para llamar a la funcion que me guarda el nombre de todos los productos con sus respectivos precios en un diccionario
productos_con_precios = obtener_nombres_y_precios(contenido)

#Accedo solo a los nombres y precios 
#Como la variable creada anteriormente es un diccionario podemos acceder a sus elementos de la siguiente forma 
nombres_de_productos = list(productos_con_precios.keys())#Accedo solo a las llaves a traves de keys
precios_de_productos = list(productos_con_precios.values())#Accedo solo a los precios a traves de values

#Creo un diccionario con las cantidad de veces que se repite el producto a partir de la cantidad de precios que tenga cada producto
frecuencia_de_productos = {}
for nombre, precios in productos_con_precios.items():
    frecuencia_de_productos[nombre] = len(precios)

#Ordenar el diccionario por cantidad de apariciones (valor)
frecuencia_ordenada = sorted(frecuencia_de_productos.items(), key=operator.itemgetter(1), reverse=True)

#Tomar los 10 más repetidos
top_10_productos = frecuencia_ordenada[:10]

#Para realizar un grafico que nos muestre esto hago lo siguiente
#Creo dos lista una para almacenar los nombres de los productos y otra para almacenar las veces que aparecen
nombres = []
cantidades = []
#Itero sobre la lista donde estan los 10 mas repetidos 
for i in top_10_productos:
    nombre = i[0]      # el primer elemento de la tupla
    cantidad = i[1]    # el segundo elemento de la tupla
    nombres.append(nombre)
    cantidades.append(cantidad)
    
#Creo un  gráfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(cantidades, labels=nombres, autopct='%1.1f%%', startangle=140)
plt.title("Distribución de los 10 productos más repetidos en mipymes")
plt.axis('equal')  # Para que el gráfico sea circular
plt.tight_layout()
plt.show()

#Calculo el total de apariciones de todos los productos 
total_apariciones = sum(cantidades)

print("Productos más repetidos y sus frecuencias:")
for nombre, cantidad in zip(nombres, cantidades):#itero sobre dos listas con zip 
    porcentaje = (cantidad / total_apariciones) * 100#Creo la formula de porciento en la que cantidad seria la parte el total de apariciones el total y para saber el % multiplico por 100
    print(f"{nombre}: {cantidad} veces ({round(porcentaje,1)}%)")#imprimo el resultado y utilizo round() para decir que me redonde el resultado 1 lugar despues de la coma



    
```

Con el análisis anterior llegué a la siguiente conclusión: Como muestra el gráfico, la mayoría de los productos que tienden a repetirse en las mipymes , suelen ser los que más necesitan los cubanos. Si se lleva esto a la realidad cubana, se entiende que todos requieren arroz, frijoles, refresco u otros alimentos para su día a día. Sin embargo, no tenía un valor exacto de lo que costaba cada uno, así que me puse a calcular el promedio de cada producto para poder decir: “Puedo ir a Artemisa con X cantidad de dinero y comprarme 1 kg de arroz.”




```python
from Funciones.biblioteca import promedio
import matplotlib.pyplot as plt
import numpy as np
# Crear un diccionario para guardar el promedio de cada uno de los 10 productos más repetidos
promedios_de_productos = {}

for nombre in nombres:  #nombres ya contiene los 10 productos más repetidos
    precios = productos_con_precios[nombre]  #Declaro una variable que va a almacenar los precios de los productos que estan en productos_con_precios
    promedio_precio = round(promedio(precios))  #Llamo a la funcion creada para calcular el promedio y lo redondeo con round
    promedios_de_productos[nombre] = promedio_precio #Agrego a mi diccionario el nombre de cada producto con el promedio de su costo 

#Para crear un grafico que nos muestre lo anterior
# Preparar datos
nombres = list(promedios_de_productos.keys())
valores = list(promedios_de_productos.values())

# Crear gráfico de barras
plt.figure(figsize=(10, 6))
bars = plt.bar(nombres, valores, color="orange")

# Añadir etiquetas con los valores encima de cada barra
for bar in bars:
    valor = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, valor + 2,
             f"{valor}", ha='center', va='bottom', fontsize=9, color="black")

plt.xticks(rotation=45, ha='right')
plt.ylabel("Precio promedio (CUP)")
plt.title("Promedio de precios de los 10 productos más repetidos en mipymes")
plt.tight_layout()
plt.show()

    
    








```

```python
from Funciones.biblioteca import extraer_valores_de_cambio
import matplotlib.pyplot as plt
from datetime import datetime
#datetime sirve para trabajar con fechas y horas de manera precisa y flexible. Es una de las librerías estándar más usadas cuando necesitas manejar tiempos, calcular diferencias, o formatear fechas

#Comence creando una variable que almacene la informacion recogida en los json de la carpeta Tasa de Cambio
direcciones_cambio = glob.glob("./Tasa de Cambio/*.json")
#Creamos una lista para guardar la informacion a traves de la cual queremos acceder cuando iteremos sobre la informacion guardada en anteriormente
contenido_cambio = []
for k in direcciones_cambio:
    contenido_cambio.append(dict(open_json(k)))
#Cree una lista para almacenar las fechas de actualizacion de la moneda en el transcurso de la investigacion
fecha = []
for f in contenido_cambio:
    fecha.append(f["Fecha de Actualizacion"])
    
#Cree una lista para guardar los diferentes precios que ha tomado la moneda mlc en el transcurso de la investigacion
valor_cambiomlc = extraer_valores_de_cambio(contenido_cambio,"MLC")

#Cree una lista para guardar los diferentes precios que ha tomado la moneda usd en el transcurso de la investigacion
valor_cambiousd = extraer_valores_de_cambio(contenido_cambio,"USD")

#Para crear un grafico que me permita visualizar lo anterior hice lo siguiente:
#Datos
fecha 
valor_cambiomlc 
valor_cambiousd 

#Converti las fechas y ordenar todo
fechas_convertidas = [datetime.strptime(f, "%d/%m/%y") for f in fecha]
datos = list(zip(fechas_convertidas, valor_cambiomlc, valor_cambiousd))
datos.sort()  # ordeno por fecha
#Aclaraciones
#datetime.strptime convierte un texto (string) que representa una fecha en un objeto de tipo datetime que Python puede entender y manipular.

#Separe los datos ya ordenados
fechas_ordenadas = [d[0] for d in datos]
mlc_ordenado = [d[1] for d in datos]
usd_ordenado = [d[2] for d in datos]
#Grafico
plt.plot(fechas_ordenadas, mlc_ordenado, 'o-b', label='MLC')
plt.plot(fechas_ordenadas, usd_ordenado, 's-g', label='USD')
plt.xticks(rotation=45)
plt.title("Comportamiento del MLC y USD en el tiempo")
plt.xlabel("Fecha de Actualización")
plt.ylabel("Valor en CUP")
plt.legend()
plt.grid(True)
#Uso las fechas originales como etiquetas
plt.xticks(ticks=fechas_convertidas, labels=fecha, rotation=45)
plt.tight_layout()
plt.show()

for fecha,mlc,usd in zip(fechas_ordenadas, mlc_ordenado, usd_ordenado):
    dia = fecha.strftime("%d/%m/%Y")  # Vonvierto las fechas en un formato legible: día/mes/año
    print(f"El día {dia} el MLC valía {mlc} CUP y el USD valía {usd} CUP")
```

Como segunda opción se tomarían las tiendas en MLC que existan en la localidad o provincia de cada ciudadano, pero aquí surge un problema: no todas las personas tienen acceso al MLC. Esto provoca que aquellos que sí lo tienen(una minoría de la población), al venderlo al cambio, puedan aumentar aún más el precio que este posee en el mercado informal. Teniendo en cuenta todas estas cosas, lo mismo ocurre con el USD, por lo que, analizando este tema, la tercera opción sería acudir a las tiendas en USD según los resultados mostrados en el gráfico.

En este punto pude comprender un poco más, ya que en la sociedad cubana, tristemente, esto es lo que se vive día a día: constantes elevaciones de precios, la presencia de un gran desbalance económico y salarios muy bajos para la adquisición de algunos productos. Llegué a la conclusión de que la situación económica cada vez se hace más difícil de comprender y asimilar, y más aún si nos guiamos por páginas como “El Toque”, en la cual se reflejan valores sin una constancia verídica que nos indique de dónde provienen. Muchas personas los utilizan para el enriquecimiento propio, sin tener en cuenta la afectación que esto genera a nivel poblacional.
