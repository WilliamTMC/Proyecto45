```python
from Funciones.biblioteca import open_json,obtener_nombres_y_precios,grafico_pastel 
import glob #Sirve para buscar archivos y directorios 
import operator#Sirve para acceder a funciones que implementan operaciones comunes (aritméticas, lógicas y de comparación) de manera programática.

#Creo una variable para almacenar todo el contenido de la carpeta Mipymes
direcciones = glob.glob("./Mipymes/*.json")#utilizo glob.glob para de la carpeta Mipymes me lea todos los .json
contenido = []
for i in direcciones :
    contenido.append(dict(open_json(i)))#llamo a la funcion open_json la cual me permite leer un json
    
#Creo una variable para llamar a la funcion (obtener_nombres_y_precios) que me guarda el nombre de todos los productos con sus respectivos precios en un diccionario 
#Parametros(lista con informacion recogida en un diccionario)
productos_con_precios = obtener_nombres_y_precios(contenido)

#Accedo solo a los nombres y precios 
nombres_de_productos = list(productos_con_precios.keys())
precios_de_productos = list(productos_con_precios.values())

#Creo un diccionario con las cantidad de veces que se repite el producto a partir de la cantidad de precios que tenga cada producto
frecuencia_de_productos = {}
for nombre, precios in productos_con_precios.items():
    frecuencia_de_productos[nombre] = len(precios)

#Ordeno los productos por frecuencia en orden descendente descendente:
frecuencia_ordenada = sorted(frecuencia_de_productos.items(),key=operator.itemgetter(1),reverse=True)

#Tomo los 10 productos con mayor frecuencia
top_10_productos = frecuencia_ordenada[:10]
#Separo nombres y cantidades para graficar 
nombres = []
cantidades = []
for i in top_10_productos: 
    nombre = i[0] 
    cantidad = i [1] 
    nombres.append(nombre)
    cantidades.append(cantidad)

#Grafico 
#Utilizo la funcion (grafico_pastel) la cual me muestra el % de repeticion de cada producto 
#Parametros(lista de valores,lista de nombres)
grafico = grafico_pastel(cantidades,nombres) 

#Muestro lo reflejado en el grafico
total_apariciones = sum(cantidades) 
print("Productos más repetidos y sus frecuencias en 30 Mipymes:")
for nombre, cantidad in zip(nombres, cantidades):
    porcentaje = (cantidad / total_apariciones) * 100#declaro una variable para saber el porciento que representa cada repeticion 
    print(f"El {nombre} se repite un total de  {cantidad} veces, lo que equivale a un ({round(porcentaje, 1)}%)")
    
```

```python
from Funciones.biblioteca import extraer_valores_de_cambio,grafico_linea
import datetime #Sirve para trabajar con fechas y horas de manera precisa y flexible. 

#Comence creando una variable que almacene la informacion recogida en los json de la carpeta Tasa de Cambio
direcciones_cambio = glob.glob("./Tasa de Cambio/*.json")#utilizo glob.glob para de la carpeta Mipymes me lea todos los .json

#Guardo la informacion anterir en una lista
#Utilizo la funcion (open_json) la cual me permite leer un json
contenido_cambio = []
for k in direcciones_cambio:
    contenido_cambio.append(dict(open_json(k)))
    
#Creo una lista para almacenar las fechas de actualizacion de la moneda en el transcurso de la investigacion
fecha = []
for f in contenido_cambio:
    fecha.append(f["Fecha de Actualizacion"])
    
#Creo dos listas para guardar los diferentes precios que ha tomado el mlc y el usd en el transcurso de la investigacion
#Utilizo la funcion (extraer_valores_de_cambio) la cual me extrae los diferentes valores que ha tomado el mlc y el usd
#Parametros(lista con el contenido de los json,nombre de la moneda(USD o MLC))
valor_cambiomlc = extraer_valores_de_cambio(contenido_cambio,"MLC")
valor_cambiousd = extraer_valores_de_cambio(contenido_cambio,"USD")

#Convierto las fechas y ordenar todo
fechas_convertidas = [datetime.datetime.strptime(f, "%d/%m/%y") for f in fecha]
datos = list(zip(fechas_convertidas, valor_cambiomlc, valor_cambiousd))
datos.sort()#Ordeno por fecha
#Aclaraciones
#datetime.strptime convierte un texto (string) que representa una fecha en un objeto de tipo datetime que Python puede entender y manipular.

#Separo los datos ya ordenados
fechas_ordenadas = [d[0] for d in datos]
mlc_ordenado = [d[1] for d in datos]
usd_ordenado = [d[2] for d in datos]

#Grafico
#Utilizo la funcion (grafico_linea) la cual me hace un grafico de linea que me permite observar el comportamientro del MLC y el USD en un rango de tiempo
#Parametros(lista de fechas,lista de valores del mlc, lista de valores del usd)
grafico4 = grafico_linea(fecha,valor_cambiomlc,valor_cambiousd)

#Muestro lo reflejado en el grafico
for fecha,mlc,usd in zip(fechas_ordenadas, mlc_ordenado, usd_ordenado):
    dia = fecha.strftime("%d/%m/%Y")#Convierto las fechas en un formato legible: dia/mes/año
    print(f"El día {dia} el MLC valía {mlc} CUP y el USD valía {usd} CUP")
```

Cada canasta fue construida considerando los precios promedio recopilados en las distintas fuentes de mercado, con el objetivo de determinar si un trabajador que percibe el salario mínimo puede acceder a ellas.Cada cifra aquí presentada es más que un número: representa la lucha diaria de las familias cubanas por acceder a lo más básico


```python
from Funciones.biblioteca import guardar_producto_lista,acceder_precio_lista,suma,grafico_precio_canastas

#Aclaraciones
salario = 2100 #Salario minimo de un trabajador en Cuba
promedios_de_productos #Nombre de productos(top 10) con sus precios al lado 

#Creo dos listas para agregar productos y precios 
canasta_basica1 = []
canasta_basica2 = []

#Productos que vamos a agregar a la primera canasta basica 
#Utilizo la funcion (guardar_producto_lista) la cual me permite guardar en una lista un producto con su precio accediendo a un diccionario
#Parametros(diccionario al cual queremos acceder,nombre del producto deseado,lista de destino)
guardar_producto_lista(promedios_de_productos,'Arroz(1kg)',canasta_basica1)
guardar_producto_lista(promedios_de_productos,'Aceite(1lt)',canasta_basica1)
guardar_producto_lista(promedios_de_productos,'Frijoles Negros(1kg)',canasta_basica1)
guardar_producto_lista(promedios_de_productos,'Jamonada de Pollo (500g)',canasta_basica1)
#Productos que vamos a agregar a la segunda canasta basica 
guardar_producto_lista(promedios_de_productos, 'Spaguetti(500g)',canasta_basica2)
guardar_producto_lista(promedios_de_productos, 'Pasta de Tomate(400g)',canasta_basica2)
guardar_producto_lista(promedios_de_productos, 'Jamonada de Pollo (500g)',canasta_basica2)

#Creo dos listas almacenar los valores de cada producto
#Utilizo la funcion (acceder_precio_lista) la cual me permite obtener el precio de los productos anterirmente guardados en listas
#Parametros(lista de productos,lista de destino)
precio_canasta1 = []
precio_canasta2 = []
acceder_precio_lista(canasta_basica1,precio_canasta1)
acceder_precio_lista(canasta_basica2,precio_canasta2)

#Para saber el precio de cada canasta sumo todos los valores 
precio_canasta1 = suma(precio_canasta1)
precio_canasta2 = suma(precio_canasta2)

#Grafico
#Utilizo la funcion (grafico_precio_canastas)  para mostrar un grafico de barras apiladas diseñado para comparar el costo de dos canastas básicas frente a un salario, mostrando claramente si hay gasto, sobrante o déficit en cada caso.
#Parametros(salario,el costo de la canasta 1 ,el costo de la canasta 2 )
grafico7 = grafico_precio_canastas(salario,precio_canasta1,precio_canasta2)

#Muestro lo reflejado en el grafico
print("Canastas Creadas")
print("Canasta Básica 1:")
for producto, precio in canasta_basica1:
    print(f"- {producto}: {precio} CUP")
print(f"Costo: ${precio_canasta1}")

print("\nCanasta Básica 2:")
for producto, precio in canasta_basica2:
    print(f"- {producto}: {precio} CUP")
print(f"Costo: ${precio_canasta2}")




```

Como puede observarse en el gráfico, se aprecia claramente la diferencia de precios entre ambas canastas. Una persona que perciba el salario mínimo tendría que optar por alternativas más simples, como es el caso de la segunda canasta básica. Sin embargo, como seres humanos, no nos gustaría consumir siempre los mismos alimentos, por lo que resulta necesario analizar cuáles de los 10 productos más frecuentes pueden ser adquiridos por un trabajador que recibe dicho salario.Este análisis abre una ventana hacia la realida mostrando la capacidad de cada familia para diversificar su dieta y mantener una alimentación digna.

```python
from Funciones.biblioteca import productos_comprables,grafico_frecuencia_compra

#Recordando
salario#Salario minimo de una persona al mes 
promedios_de_productos#Productos y precios 

#Creo una variable para almacenar los productos y precios
#Utilizo la funcion (productos_comprables) la cual va a sacarme de un diccionario los productos que tengan un precio menor o igual a x salario y los va a agregar a una lista con su precio y la cantidad de veces que puedo comprarlo
#Parametros(diccionario, un entero x )
productos_vs_salario = productos_comprables(promedios_de_productos,salario)
    
#Separo los datos
productos = [p[0] for p in productos_vs_salario]#nombres
cantidades = [p[2] for p in productos_vs_salario]#valores

#Grafico
#Utilizo la funcion (grafico_frecuencia_compra) la cual me muestra un grafico de puntos  en el que puedo oservar la cantidad de veces que puedo comprar un producto
#Parametros(lista con nombre de los productos,lista con la frecuencia de compras)
grafico8 = grafico_frecuencia_compra(productos,cantidades)

#Muestro lo reflejado en el grafico
for producto, precio, cantidad in productos_vs_salario:
    print(f"- Puedes comprar {producto} a ${precio} CUP  {cantidad} veces")  
```

Como refleja el gráfico, estos serían los productos que una persona podría comprar en Cuba con un salario mínimo de 2 100 CUP. A simple vista, podemos observar que, para una persona con ingresos tan bajos, adquirir un paquete de pollo implicaría ahorrar el equivalente a dos meses de trabajo.

Esto refleja la realidad que enfrentan miles de familias cubanas. La imposibilidad de acceder a un alimento tan básico como el pollo muestra la magnitud de la brecha entre ingresos y necesidades.
