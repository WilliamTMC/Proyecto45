```python
import matplotlib.pyplot as plt
import numpy as np
#Declaramos dos variables para almacenar la informacion de los json que guardan la informacion sobre las tiendas en mlc y usd 
tienda_mlc = open_json("Tiendas en MlC y USD/mlc.json")
tienda_usd = open_json("Tiendas en MlC y USD/usd.json")
#En esta variable tengo los productos mas repetidos entre mipymes que es a traves de los cuales estamos haciendo el analisis
productos
#Creamos dos lista una para almacenar los nombres y otra para almacenar los precios de los productos mas repetidos en las mipymes que se encuentren en las tienda de mlc 
nombre_productosmlc = []
precio_productosmlc = []
for mlc in productos:
    for m in tienda_mlc["Productos"]:
        if m["nombre"] == mlc:
            nombre_productosmlc.append(m['nombre'])
            precio_productosmlc.append(m['precio'])
#Hago lo mismo con la tienda en usd 
nombre_productosusd = []
precio_productosusd = []
for usd in productos:
    for u in tienda_mlc["Productos"]:
        if u["nombre"] == usd:
            nombre_productosusd.append(u['nombre'])
            precio_productosusd.append(u['precio'])
#Para crear un grafico que nos muestre los resultados de ambas tiendas
#Datos
# Listas con productos y precios
nombre_productosmlc 
precio_productosmlc
nombre_productosusd 
precio_productosusd 

#Creo una figura con dos subplots horizontales
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
#fig es el objeto principal sobre el cual voy a trabajar
#ax1 y ax2 representan los graficos o sea ax1 seria el grafico de la izquierda y ax2 seria el grafico de la derecha
#Aclaraciones
#1-plt.subplots crea una figura y uno o varios ejes (subgráficos) dentro de ella
#2-(1, 2): significa que quieres 1 fila y 2 columnas de subgráficos → es decir, dos gráficos lado a lado.
#3-figsize=(14, 6): define el tamaño total de la figura en pulgadas (14 de ancho × 6 de alto).


#Grafico MLC
x1 = np.arange(len(nombre_productosmlc))
x2 = np.arange(len(nombre_productosusd))
ax1.bar(x1, precio_productosmlc, color='skyblue')
ax1.set_title("Precios en Tienda MLC")
ax1.set_xlabel("Nombre de productos")
ax1.set_ylabel("Precio en MLC")
ax1.set_xticks(x1)
ax1.set_xticklabels(nombre_productosmlc, rotation=45, ha='right')

#Grafico USD
ax2.bar(x2, precio_productosusd, color='lightgreen')
ax2.set_title("Precios en Tienda USD")
ax2.set_xlabel("Nombre de productos")
ax2.set_ylabel("Precio en USD")
ax2.set_xticks(x2)
ax2.set_xticklabels(nombre_productosusd, rotation=45, ha='right')

#Ajusto el diseño
plt.tight_layout()
plt.show()

#Para mostrar precios debajo
print("Precios en Tienda MLC:")
for nombre,precio in zip(nombre_productosmlc, precio_productosmlc):
    print(f"El producto {nombre} vale ${precio} MLC")

print("\nPrecios en Tienda USD:")
for nombre,precio in zip(nombre_productosusd, precio_productosusd):
    print(f"El producto {nombre} vale ${precio} USD")


        
```

```python
import matplotlib.pyplot as plt
from datetime import datetime
#datetime sirve para trabajar con fechas y horas de manera precisa y flexible. Es una de las librerías estándar más usadas cuando necesitas manejar tiempos, calcular diferencias, o formatear fechas

#Comence creando una variable que almacene la informacion recogida en los json de la carpeta Tasa de Cambio
direcciones_cambio = glob.glob("./Tasa de Cambio/*.json")
#Creamos una lista para guardar la informacion a traves de la cual queremos acceder cuando iteremos sobre la informacion guardada en anteriormente
contenido_cambio = []
for k in direcciones_cambio:
    contenido_cambio.append(dict(open_json(k)))
#Cree una lista para almacenar las fechas de actualizacion de la moneda en el transcurso de la investigacion
fecha = []
for f in contenido_cambio:
    fecha.append(f["Fecha de Actualizacion"])
#Cree una lista para guardar los diferentes precios que ha tomado la moneda mlc en el transcurso de la investigacion
valor_cambiomlc = []
for preciomlc in contenido_cambio:
    for vprecio in preciomlc["Cambio"]:
        if vprecio["nombre"] == "MLC":
            valor_cambiomlc.append(vprecio["valor"])
#Cree una lista para guardar los diferentes precios que ha tomado la moneda usd en el transcurso de la investigacion
valor_cambiousd = []
for preciousd in contenido_cambio:
    for vprecio in preciousd["Cambio"]:
        if vprecio["nombre"] == "USD":
            valor_cambiousd.append(vprecio["valor"])
#Para crear un grafico que me permita visualizar lo anterior hice lo siguiente:
#Datos
fecha 
valor_cambiomlc 
valor_cambiousd 

#Converti las fechas y ordenar todo
fechas_convertidas = [datetime.strptime(f, "%d/%m/%y") for f in fecha]
datos = list(zip(fechas_convertidas, valor_cambiomlc, valor_cambiousd))
datos.sort()  # ordeno por fecha
#Aclaraciones
#datetime.strptime convierte un texto (string) que representa una fecha en un objeto de tipo datetime que Python puede entender y manipular.

#Separe los datos ya ordenados
fechas_ordenadas = [d[0] for d in datos]
mlc_ordenado = [d[1] for d in datos]
usd_ordenado = [d[2] for d in datos]

#Grafico
plt.plot(fechas_ordenadas, mlc_ordenado, 'o-b', label='MLC')
plt.plot(fechas_ordenadas, usd_ordenado, 's-g', label='USD')
plt.xticks(rotation=45)
plt.title("Comportamiento del MLC y USD en el tiempo")
plt.xlabel("Fecha de Actualización")
plt.ylabel("Valor en CUP")
plt.legend()
plt.grid(True)
#Uso las fechas originales como etiquetas
plt.xticks(ticks=fechas_convertidas, labels=fecha, rotation=45)
plt.tight_layout()
plt.show()

for fecha,mlc,usd in zip(fechas_ordenadas, mlc_ordenado, usd_ordenado):
    dia = fecha.strftime("%d/%m/%Y")  # Vonvierto las fechas en un formato legible: día/mes/año
    print(f"El día {dia} el MLC valía {mlc} CUP y el USD valía {usd} CUP")
```

Como pude observar, noté que a medida que pasaban los días, el MLC, que comenzó con un precio inferior al USD, empezó a aumentar. El USD, en cambio, parecía inestable: lo mismo subía que bajaba.
